```python
#Data and data frame creation
import pandas as pd
data = {
    'Name' : ['Amal', 'Kamal', 'Wimal', 'Kalum', 'Almeda', 'Narmadha', 'Bandara', 'Vipula', 'Sujatha', 'Sumana'],
    'IT2010' : [70, 90, 34, 67, 89, 89, 56, 31, 98, 10],
    'IT2020' : [25, 45, 44, 31, 89, 90, 67, 77, 91, 45],
    'IT2030' : [44, 56, 32, 43, 89, 76, 67, 45, 11, 45],
    'IT2040' : [99, 99, 98, 91, 78, 76, 43, 32, 56, 78],
    'IT2050' : [66, 77, 86, 86, 90, 90, 43, 34, 12, 32],
  }
df = pd.DataFrame(data)
df
```

```python
#Five Number Summary, mead and sd
df.describe()
```

```python
#Bar Graph Generation
import matplotlib.pyplot as plt

df.set_index('Name').plot(kind='bar')
plt.title("Student Marks Comparison")
plt.ylabel("Marks")
plt.show()
```

```python
#Calculate the Total and Average Marks of students
df['Total'] = df['IT2010'] + df['IT2020'] + df['IT2030'] + df['IT2040'] + df['IT2050']
df['Average'] = df['Total'] / 5

df
```

```python
#Find the top student using average marks
top_student = df.loc[df['Average'].idxmax()]
top_student
```

```python
#Ranking the students based on average marking
df['Rank'] = df['Average'].rank(ascending = False)
#Pass or Fail column
df['Result'] = df['Average'].apply(lambda x: 'Pass' if x >= 45 else 'Fail')
df.sort_values('Rank')

```

```python
df.plot(x = 'Name', y = 'Average', kind = 'bar')
```
